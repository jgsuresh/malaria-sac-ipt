```python
import pandas as pd
import numpy as np
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("talk")
sns.set_style("white")
```

```python
df1 = pd.read_csv("sim_data_0c36f501-afd1-ee11-aa11-b88303911bc1.csv") # Southern/Central
df2 = pd.read_csv("sim_data_d6f3a939-bfd1-ee11-aa11-b88303911bc1.csv") # Sahel

df = pd.concat([df1,df2], ignore_index=True)
```

```python
foo = df.groupby(["archetype", "baseline_transmission_metric", "transmission_level", "scenario_number"]).agg("mean").reset_index()
foo2 = df.groupby(["archetype", "baseline_transmission_metric", "transmission_level", "scenario_number"]).agg("std").reset_index()
foo3 = pd.merge(foo,foo2, suffixes=('','_STDEV'), on=["archetype", "baseline_transmission_metric", "transmission_level", "scenario_number"])
```

```python
base = foo3[foo3["scenario_number"]==0]
base = base[["archetype",
             "baseline_transmission_metric",
             "transmission_level",
             "cases_treated",
             'clinical_incidence0_5',
             'clinical_incidence2_10',
             'clinical_incidence6_15',
             'clinical_incidence16_500',
             'clinical_incidence_all',
             'pfpr0_5',
             'pfpr2_10',
             'pfpr6_15',
             'pfpr16_500',
             'pfpr_all']]
base
```

```python
bar = pd.merge(df, base, how="left", on=["archetype", "baseline_transmission_metric", "transmission_level"], suffixes=("","_BASELINE"))
bar
```

```python
# Add total clinical burden fields:
case_type = "clinical"
for agebin in ["0_5", "2_10","6_15","16_500","_all"]:
    # for case_type in ["clinical", "severe"]:
    old = bar["{}_incidence{}_BASELINE".format(case_type, agebin)]
    new = bar["{}_incidence{}".format(case_type, agebin)]
    frac_averted = (old-new)/old

    bar["raw_averted_{}_{}".format(case_type, agebin)] = old-new
    bar["frac_averted_{}_{}".format(case_type, agebin)] = (old-new)/old
    
    # errorbar
    # new_std = bar["{}_incidence{}_STDEV".format(case_type, agebin)]
    # frac_averted_stdev = new_std/old
    
    # bar["raw_averted_{}_{}_STDEV".format(case_type, agebin)] = new_std
    # bar["frac_averted_{}_{}_STDEV".format(case_type, agebin)] = frac_averted_stdev
```

```python
a = bar.groupby(["scenario_number"]).agg("mean").reset_index()
a_low = bar.groupby(["scenario_number"]).agg(lambda x: np.percentile(x,5)).reset_index()
a_high = bar.groupby(["scenario_number"]).agg(lambda x: np.percentile(x,95)).reset_index()
```

OLD: When long-acting drugs are administered frequently, the model predicts 70-90% of clinical cases in school-age children are averted across all transmission archetypes and levels, and 20-60% of clinical cases are averted in younger children and adults, with more cases averted where transmission levels are lower (PfPR2-10 5-10%). Shorter-acting drugs administered monthly and longer-acting drugs administered once per school-term averted 40-60% of cases in school-age children and 15-50% of cases in younger children and adults. From the health system perspective, the cost of to implement implementing IPTsc may be partially or completely offset by the savings accrued by reduced clinical burden on the health system.
