```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Sample both costs and cases in a way that preserves the correlation between the two.
df = pd.read_csv("sim_data_raw_with_cost_ranges.csv")
df
```

```python
df["cases"] = df["clinical_incidence_all"]*5000
```

```python
def sample_as_normal(min_val, max_val, n_samples):
    # Sample from normal distribution where 99% of probability distribution is within min and max
    mean = (min_val + max_val) / 2
    sigma = 0.194*(max_val - min_val)
    return np.random.normal(mean, sigma, n_samples)
```

```python
foo = df[(df["archetype"]=="Southern") & (df["transmission_level"]==0.2) & (df["scenario_number"]==6)]
foo
```

```python
costs = []

indices = np.random.choice(np.arange(100), 10000)

# run_numbers = np.random.choice(foo["Run_Number"], 1000)
run_numbers = foo["Run_Number"].values[indices]
cases = foo["cases"].values[indices]

# For each of these run_numbers, get the corresponding cases. OK if there's repetition
# cases = 
# cases = np.array(foo[foo["Run_Number"].isin(run_numbers)]["cases"])

for i in range(10000):
    run_number = run_numbers[i]
    # Sample cost from normal distribution for this run number
    cost_low = foo[foo["Run_Number"]==run_number]["cost_low"].iloc[0]
    cost_high = foo[foo["Run_Number"]==run_number]["cost_high"].iloc[0]
    cost = sample_as_normal(cost_low, cost_high, 1)
    costs.append(cost[0])
```

```python
plt.scatter(cases, costs)
```

```python
# Correlation between 
correlation_matrix = np.corrcoef(cases, costs)
correlation_matrix
```

```python
def get_costs_and_cases(archetype="Southern", transmission_level=0.2, scenario_number=0):
    foo = df[(df["archetype"]==archetype) & (df["transmission_level"]==transmission_level) & (df["scenario_number"]==scenario_number)]
    costs = []
    indices = np.random.choice(np.arange(100), 10000)
    run_numbers = foo["Run_Number"].values[indices]
    cases = foo["cases"].values[indices]
    for i in range(10000):
        run_number = run_numbers[i]
        cost_low = foo[foo["Run_Number"]==run_number]["cost_low"].iloc[0]
        cost_high = foo[foo["Run_Number"]==run_number]["cost_high"].iloc[0]
        cost = sample_as_normal(cost_low, cost_high, 1)
        costs.append(cost[0])
    return np.array(cases), np.array(costs)
```

```python
ca0, co0 = get_costs_and_cases(scenario_number=0)
ca6, co6 = get_costs_and_cases(scenario_number=6) # asaq term
ca7, co7 = get_costs_and_cases(scenario_number=7) # asaq monthly
ca8, co8 = get_costs_and_cases(scenario_number=8) # dp term
ca9, co9 = get_costs_and_cases(scenario_number=9) # dp monthly
```

```python
# ica = ca6-ca0
# ico = co6-co0
# ica = ca7-ca6
# ico = co7-co6
ica = ca7-ca8
ico = co7-co8
```

```python
plt.scatter(ica, ico)
print(np.corrcoef(ica, ico))
print("Prob of less effective and less costly: ", np.mean((ica>0) & (ico<0)))
print("Prob of more effective and more costly: ", np.mean((ica<0) & (ico>0)))
print("Prob of less effective and more costly: ", np.mean((ica>0) & (ico>0)))
print("Prob of more effective and less costly: ", np.mean((ica<0) & (ico<0)))
```
