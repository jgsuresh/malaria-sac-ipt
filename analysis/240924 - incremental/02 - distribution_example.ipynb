```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Let A, B, and C be uniformly distributed random variables between 0 and 1. Show distribution of A+B+C
n = 10000
A = np.random.rand(n)
B = np.random.rand(n)
C = np.random.rand(n)
D = A + B + C
sns.histplot(D, bins=100, kde=True)

# Add lines showing the mean and standard deviation
mean = np.mean(D)
std = np.std(D)
plt.axvline(mean, color='r', linestyle='dashed', linewidth=1)
plt.axvline(mean - std, color='g', linestyle='dashed', linewidth=1)
plt.axvline(mean + std, color='g', linestyle='dashed', linewidth=1)

# Add lines for 95% confidence interval
ci = 1.96 * std
print(ci)
plt.axvline(mean - ci, color='b', linestyle='dashed', linewidth=1)
plt.axvline(mean + ci, color='b', linestyle='dashed', linewidth=1)

plt.show()
```

```python
# Let A, B, and C be uniformly distributed random variables between 0 and 1. Show distribution of A+B+C
n = 10000
A = np.random.normal(0.5, 0.194, n)
B = np.random.normal(0.5, 0.194, n)
C = np.random.normal(0.5, 0.194, n)
D = A + B + C
sns.histplot(D, bins=100, kde=True)

# Add lines showing the mean and standard deviation
mean = np.mean(D)
std = np.std(D)
plt.axvline(mean, color='r', linestyle='dashed', linewidth=1)
plt.axvline(mean - std, color='g', linestyle='dashed', linewidth=1)
plt.axvline(mean + std, color='g', linestyle='dashed', linewidth=1)

# Add lines for 95% confidence interval
ci = 1.96 * std
print(ci)
plt.axvline(mean - ci, color='b', linestyle='dashed', linewidth=1)
plt.axvline(mean + ci, color='b', linestyle='dashed', linewidth=1)

plt.show()
```

```python
def sample_as_normal(min, max, n_samples):
    # Sample from normal distribution where 99% of probability distribution is within min and max
    mean = (min + max) / 2
    sigma = 0.194*(max - min)
    return np.random.normal(mean, sigma, n_samples)
```

```python
A = sample_as_normal(2, 10, int(1e6))
```

```python
plt.hist(A, bins=100)
```
