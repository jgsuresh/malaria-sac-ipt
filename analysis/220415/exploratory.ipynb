```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_context("talk")
sns.set_style("white")
```

```python
df_sahel_scenarios = df_scenarios[df_scenarios["archetype"]=="Sahel"].reset_index(drop=True)
```

```python
df_all = df_sim.merge(df_sahel_scenarios, how="left", on="scenario_number")
df_all
```

```python
df_base = df_all[df_all["scenario_number"]==30].groupby("baseline_eir").agg("mean").reset_index()
df_base
```

```python
plt.plot(df_base["baseline_eir"], df_base["clinical_incidence0_5"], label="0-5")
plt.plot(df_base["baseline_eir"], df_base["clinical_incidence6_15"], label="6-15")
plt.plot(df_base["baseline_eir"], df_base["clinical_incidence16_500"], label="16+")
plt.plot(df_base["baseline_eir"], df_base["clinical_incidence_all"], label="All")
plt.xlabel("baseline_eir")
plt.ylabel("clinical_incidence")
plt.legend()
```

```python
plt.plot(df_base["baseline_eir"], df_base["pfpr0_5"], label="0-5")
plt.plot(df_base["baseline_eir"], df_base["pfpr6_15"], label="6-15")
plt.plot(df_base["baseline_eir"], df_base["pfpr16_500"], label="16+")
plt.plot(df_base["baseline_eir"], df_base["pfpr_all"], label="All")
plt.xlabel("baseline eir")
plt.ylabel("pfpr")
plt.legend()
```

```python
df_agg = df_all.groupby(["baseline_eir", "scenario_number"]).agg("mean").reset_index()
df_agg = df_agg.merge(df_base, on="baseline_eir", suffixes=["","_BASELINE"], how="left")
df_agg
```

```python
# Add total clinical burden fields:
for agebin in ["0_5", "2_10","6_15","16_500","_all"]:
    for case_type in ["clinical", "severe"]:
        old = df_agg["{}_incidence{}_BASELINE".format(case_type, agebin)]
        new = df_agg["{}_incidence{}".format(case_type, agebin)]
        frac_averted = (old-new)/old
        
        df_agg["frac_averted_{}_{}".format(case_type, agebin)] = frac_averted
```

```python
df_agg
```

```python
columns_to_keep = ["scenario_number", "baseline_eir"]
for agebin in ["0_5", "2_10","6_15","16_500","_all"]:
    for case_type in ["clinical", "severe"]:
        columns_to_keep += [f"{case_type}_incidence{agebin}"]
        columns_to_keep += [f"frac_averted_{case_type}_{agebin}"]
df_plot = df_agg[columns_to_keep]
df_plot.sort_values(by=["scenario_number", "baseline_eir"], inplace=True)
df_plot
```

```python
label_dict = {
    13: "Presumptive DP each term",
    22: "Screen-and-treat DP each month",
    16: "Presumptive DP each month",
}

scenario_list = list(label_dict.keys())

s_dict = {}
for s in scenario_list:
    s_dict[s] = df_agg[df_agg["scenario_number"]==s]

    
color_dict = {}
for i in range(3):
    color_dict[scenario_list[i]] = "C{}".format(i)
    
    
plt.figure(dpi=300, figsize=(8,5))

    
agebins = ["0_5", "6_15", "16_500"]
agebin_labels = ["Ages 0-5", "Ages 6-15", "Ages 16+"]

for i in range(3):
    agebin = agebins[i]
    agebin_label = agebin_labels[i]
    plt.subplot(1,3,i+1)
    
    for s in scenario_list:
        linestyle='solid'
        if color_dict:
            c = color_dict[s]

        x = s_dict[s]["baseline_eir"]
        y = s_dict[s]["frac_averted_clinical_{}".format(agebin)]
        plt.plot(x,y, label=label_dict[s], linestyle=linestyle, c=c)
        if i == 2:
            plt.legend(fontsize=6)
        if i == 1:
            plt.xlabel("Baseline EIR")
        if i == 0:
            plt.ylabel("Fraction of clinical burden averted")
        plt.ylim([0,1])
#         plt.xlabel("Baseline PfPR2-10")
        plt.title(agebin_label)
        plt.xscale("log")
plt.suptitle("Clinical burden averted")
plt.tight_layout()
```
