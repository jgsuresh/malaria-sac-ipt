```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_context("talk")
sns.set_style("white")
```

# Summarize main scenario simulation output

```python
# df1 = pd.read_csv("sim_data_12818e17-5029-ed11-a9fc-b88303911bc1.csv")
# df2 = pd.read_csv("sim_data_1b2a31b5-5129-ed11-a9fc-b88303911bc1.csv")

df1 = pd.read_csv("sim_data_f6f2b8f4-dc32-ed11-a9fc-b88303911bc1.csv")
df2 = pd.read_csv("sim_data_bf038501-a7d0-ec11-a9f8-b88303911bc1.csv")
df = pd.concat([df1,df2], ignore_index=True)
```

# Get baseline for comparisons

```python
foo = df.groupby(["archetype", "baseline_eir", "scenario_number"]).agg("mean").reset_index()
foo2 = df.groupby(["archetype", "baseline_eir", "scenario_number"]).agg("std").reset_index()
foo3 = pd.merge(foo,foo2, suffixes=('','_STDEV'), on=["archetype", "baseline_eir", "scenario_number"])
```

```python
base = foo3[foo3["scenario_number"]==30]
base = base[["archetype", 
             "baseline_eir", 
             "cases_treated", 
             'clinical_incidence0_5', 
             'clinical_incidence2_10', 
             'clinical_incidence6_15', 
             'clinical_incidence16_500',
             'clinical_incidence_all',
            'pfpr0_5',
            'pfpr2_10',
            'pfpr6_15',
             'pfpr16_500',
            'pfpr_all']]
base
```

```python
bar = pd.merge(foo3, base, how="left", on=["archetype", "baseline_eir"], suffixes=("","_BASELINE"))
bar
```

```python
# Add total clinical burden fields:
case_type = "clinical"
for agebin in ["0_5", "2_10","6_15","16_500","_all"]:
    # for case_type in ["clinical", "severe"]:
    old = bar["{}_incidence{}_BASELINE".format(case_type, agebin)]
    new = bar["{}_incidence{}".format(case_type, agebin)]
    frac_averted = (old-new)/old

    bar["frac_averted_{}_{}".format(case_type, agebin)] = frac_averted
    
    # errorbar
    new_std = bar["{}_incidence{}_STDEV".format(case_type, agebin)]
    frac_averted_stdev = new_std/old
    
    bar["frac_averted_{}_{}_STDEV".format(case_type, agebin)] = frac_averted_stdev
```

```python
# Add prevalence reduction field:
for agebin in ["0_5", "2_10","6_15","16_500","_all"]:
    # for case_type in ["clinical", "severe"]:
    old = bar["pfpr{}_BASELINE".format(agebin)]
    new = bar["pfpr{}".format(agebin)]
    frac_averted = (old-new)/old

    bar["frac_averted_{}_{}_PFPR".format(case_type, agebin)] = frac_averted
    
    
    # errorbar
    new_std = bar["pfpr{}_STDEV".format(agebin)]
    frac_averted_stdev = new_std/old
    
    bar["frac_averted_{}_{}_PFPR_STDEV".format(case_type, agebin)] = frac_averted_stdev
```
