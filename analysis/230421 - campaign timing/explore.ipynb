Revisiting general problem: output all possible campaign timings, given N total campaigns, whether you are limited to school year, and whether you are limited to M per term.

```python
def check_elements(A, B):
    return all(element in B for element in A)
```

```python
def generate_lists(min_value, max_value, min_diff, list_length, school_calendar=None, days_divisible_by=7):
    if school_calendar is None:
        school_calendar = list(range(1,365+1))
        
    result = []

    def generate_recursive(prefix, start, remaining):
        if remaining == 0:
            if check_elements(prefix, school_calendar):
                result.append(prefix)
            return
        for i in range(start, max_value - (min_diff * (remaining - 1)) + 1, days_divisible_by):
            generate_recursive(prefix + [i], i + min_diff, remaining - 1)

    generate_recursive([], min_value, list_length)
    return result
```

Note that I force only days every 2 weeks to be viable, since allowing every week with 6 campaigns blows up (~200k options).  This should be fine for the Central archetype, where exact campaign timings are not as critical

```python
min_value = 1
max_value = 365
min_diff = 28
list_length = 6
days_divisible_by = 14

# Ensure min_value is divisible by 7
min_value = (min_value + 6) // days_divisible_by * days_divisible_by

lists = generate_lists(min_value, max_value, min_diff, list_length, school_calendar=central_days, days_divisible_by=days_divisible_by)

print(len(lists))
df = pd.DataFrame(lists)
df.to_csv("central_6_campaigns.csv", index=False)
```
