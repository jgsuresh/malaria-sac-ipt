```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_context("talk")
sns.set_style("white")
```

```python
df = pd.read_csv("sim_data_full_220708.csv")
```

```python
foo = df.groupby(["archetype", "baseline_eir", "scenario_number"]).agg("mean").reset_index()
foo
```

```python
foo.columns
```

```python
base = foo[foo["scenario_number"]==30]
base = base[["archetype", 
             "baseline_eir", 
             "cases_treated", 
             'clinical_incidence0_5', 
             'clinical_incidence2_10', 
             'clinical_incidence6_15', 
             'clinical_incidence16_500',
             'clinical_incidence_all',
            'pfpr0_5',
            'pfpr2_10',
            'pfpr6_15',
             'pfpr16_500',
            'pfpr_all']]
base
```

```python
bar = pd.merge(foo, base, how="left", on=["archetype", "baseline_eir"], suffixes=("","_BASELINE"))
bar
```

```python
# Add total clinical burden fields:
case_type = "clinical"
for agebin in ["0_5", "2_10","6_15","16_500","_all"]:
    # for case_type in ["clinical", "severe"]:
    old = bar["{}_incidence{}_BASELINE".format(case_type, agebin)]
    new = bar["{}_incidence{}".format(case_type, agebin)]
    frac_averted = (old-new)/old

    bar["frac_averted_{}_{}".format(case_type, agebin)] = frac_averted
```

```python
# Add prevalence reduction field:
for agebin in ["0_5", "2_10","6_15","16_500","_all"]:
    # for case_type in ["clinical", "severe"]:
    old = bar["pfpr{}_BASELINE".format(agebin)]
    new = bar["pfpr{}".format(agebin)]
    frac_averted = (old-new)/old

    bar["frac_averted_{}_{}_PFPR".format(case_type, agebin)] = frac_averted
```

```python
# df_cost = pd.read_csv("marita_results_2022-07-08.csv") # incorrect CHW costs
# df_cost = pd.read_csv("marita_results_2022-07-15.csv") # only single CHW overhead cost, but want to be able to see if this can be reduced
# df_cost.drop(["Unnamed: 0", "drug_type", "Scenario", "cases"], axis=1, inplace=True)
df_cost = pd.read_csv("marita_results_2022-07-21.csv")
df_cost.drop(["Unnamed: 0", "drug_type", "Scenario"], axis=1, inplace=True)
df_cost
```

```python
df_save = pd.merge(bar, df_cost, how="left", on=["archetype", "baseline_eir", "scenario_number"])
df_save
```

```python
df_scenarios = pd.read_csv("../../run_sims/Assets/scenario_master_list.csv")
df = pd.merge(df_save, df_scenarios, on=["archetype", "scenario_number"])
df
```
