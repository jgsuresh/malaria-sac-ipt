```python
import pandas as pd
import numpy as np
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("talk")
sns.set_style("white")
```

```python
df = pd.read_csv("sim_data_summary_220715.csv")
df
```

```python
scenarios_to_plot = [1,4,7,10,13,16,19,22]
df_scenarios = df[np.in1d(df["scenario_number"], scenarios_to_plot)].reset_index(drop=True)
df_scenarios
```

```python
df_scenarios.columns
```

```python
for a, sdf in df_scenarios.groupby(["archetype", "baseline_eir"]):
    plt.figure(dpi=300)
    sns.scatterplot(data=sdf, x="frac_averted_clinical__all", y="cost", style="drug_type", hue="screen_type", size="interval", size_order=["month","term"])
    plt.title(f"{a[0]} - eir {a[1]}") 
    plt.legend(fontsize="xx-small")
```
