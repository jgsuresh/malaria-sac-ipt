```python
def create_stacked_barplot_df(sdf):
    # sdf should be single archetype and single baseline_eir
    
    archetype = sdf["archetype"].iloc[0]
    
    column_list = ["campaign_type", 'cost.smc', 'cost.bednet', 'cost.HF', 'cost.CHW', 'cost.RDT_IPTsc', 'cost.drug']
    if archetype != "Sahel":
        column_list.pop(1)
        
        
    to_plot = sdf[column_list]
    for c in column_list:
        if c == "campaign_type":
            pass
        else:
            to_plot[c] = to_plot[c]/5 # make it cost per 1000
    
    
    to_plot.rename(columns={"cost.smc": "SMC",
                                            "cost.bednet": "ITNs",
                                            "cost.HF": "Health facility treatment",
                                            "cost.CHW": "CHW cost (IPTsc)",
                                            "cost.RDT_IPTsc": "RDT cost (IPTsc)",
                                            "cost.drug": "Drug cost (IPTsc)"},
                  inplace=True)
    

    
    color_dict = {"ITNs": "C0", "Health facility treatment": "C1", "CHW cost (IPTsc)": "C2", "RDT cost (IPTsc)": "C3", "Drug cost (IPTsc)": "C4", "SMC": "C5"}
    
    plt.figure(dpi=300)
    to_plot.set_index("campaign_type").plot(kind='bar', stacked=True, linewidth=0, color=color_dict)
    plt.legend(fontsize=8)
    plt.ylabel("Cost per 1000 in population")
    
    return to_plot
    
```

```python
sdf
```

```python
sdf = df_bar[np.logical_and(df_bar["archetype"]=="Central", df_bar["baseline_eir"]==10)]
b = create_stacked_barplot_df(sdf)
```

```python
sdf = df_bar[np.logical_and(df_bar["archetype"]=="Central", df_bar["baseline_eir"]==10)]
create_stacked_barplot_df(sdf)
```

```python
sdf = df_bar[np.logical_and(df_bar["archetype"]=="Southern", df_bar["baseline_eir"]==10)]
create_stacked_barplot_df(sdf)
```
