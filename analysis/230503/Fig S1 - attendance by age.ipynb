```python
import pandas as pd
import numpy as np
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("talk")
sns.set_style("white")
```

```python
df = pd.read_csv("../../run_sims/Assets/primary_school_attendance_by_age.csv")
```

```python
plt.figure(dpi=300)
plt.plot(df["age"], df["relative_attendance"]*0.8, marker='o')
plt.xlabel("Age")
plt.ylabel("Percentage attending primary school")
plt.xticks(range(6,19))
plt.ylim([0,1])
plt.savefig("age_distribution.png")
plt.savefig("age_distribution.pdf")
```
