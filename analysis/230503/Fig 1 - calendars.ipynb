```python
import pandas as pd
import numpy as np
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("talk")
sns.set_style("white")
```

```python
df = pd.read_csv("incidence_curves_smoothed.csv")
```

```python
for a, sdf in df.groupby("archetype"):
    plt.plot(sdf["day"], sdf["cases"],label=a)
plt.legend()
plt.ylabel("Daily cases")
plt.xlabel("Day")
```

```python
month_days = 30.417/2. + np.array([    0.0,
    30.417,
    60.833,
    91.25,
    121.667,
    152.083,
    182.5,
    212.917,
    243.333,
    273.75,
    304.167,
    334.583])
```

```python
# previous colors
southern_color = "#7B5DA6"
sahel_color = "#E5537E"
cent_color = "#2B8B89"
```

```python
school_calendar_color = "#009F89" #D51C5B" #"#5292C2"
school_campaign_color = "#009F89"
```

```python
# School calendars:
archetype = "Southern"
df_plot = df[df["archetype"]=="Southern"]
plt.plot(df_plot["day"], df_plot["cases"], color='#262626')

plt.axvspan(15,105, alpha=0.2, color=school_calendar_color, linewidth=0)
plt.axvspan(135,227, alpha=0.2, color=school_calendar_color, linewidth=0)
plt.axvspan(244,335, alpha=0.2, color=school_calendar_color, linewidth=0)
plt.xlim([0,365])

plt.xticks(month_days)
ax = plt.gca()
ax.set_xticklabels(["J", "F", "M", "A", "M", "J", "J", "A", "S", "O", "N", "D"])

for td in [15,165,254]:
    plt.axvline(td, color=school_campaign_color)
    
for md in [15,46,74,105,135,166,196,227,244,274,305,335]:
    plt.axvline(md,color=school_campaign_color, lw=1, zorder=5, linestyle='dashed')

plt.ylim([0,40])
ax.set_yticks([])
plt.title("Southern archetype")
# plt.savefig("calendar_southern.pdf")
```
