# Summarize main scenario simulation output

```python
df1 = pd.read_csv("sim_data_643e3a04-cb7c-ed11-aa00-b88303911bc1.csv") # Sahel
df2 = pd.read_csv("sim_data_7acf4b3f-cf7c-ed11-aa00-b88303911bc1.csv") #south/central
df3 = pd.read_csv("sim_data_e5b8cd59-de7c-ed11-aa00-b88303911bc1.csv") #s/c more seeds
df4 = pd.read_csv("sim_data_02c497c9-007d-ed11-aa00-b88303911bc1.csv") #s/c more seeds


df10 = pd.read_csv("../221207/sim_data_6687eab5-6867-ed11-a9ff-b88303911bc1.csv") # Southern/Central, pfpr, new ASAQ, health-seeking sweep 1/2
df11 = pd.read_csv("../221207/sim_data_600fc461-d266-ed11-a9ff-b88303911bc1.csv") # Southern/Central, pfpr, new ASAQ, health-seeking sweep 2/2
df12 = pd.read_csv("../221207/sim_data_fac14d60-9d76-ed11-aa00-b88303911bc1.csv") #Sahel, pfpr, new ASAQ, health-seeking, 100 seeds
df13 = pd.read_csv("../221207/sim_data_8a020b38-1777-ed11-aa00-b88303911bc1.csv") #Sahel, pfpr, new ASAQ, health-seeking, 100 MORE seeds
df14 = pd.read_csv("../221207/sim_data_1de39b01-5b77-ed11-aa00-b88303911bc1.csv") #Sahel, pfpr, new ASAQ, health-seeking, 100 MORE MORE seeds
df15 = pd.read_csv("../221207/sim_data_9d7d39c6-3377-ed11-aa00-b88303911bc1.csv")
df16 = pd.read_csv("../221207/sim_data_0e6fedec-3d77-ed11-aa00-b88303911bc1.csv")

df20 = pd.read_csv("sim_data_bdd2b5f7-6d7d-ed11-aa00-b88303911bc1.csv") #all archetypes, IST

df = pd.concat([df1,df2,df3,df10,df11,df12,df13,df14,df15,df16, df20], ignore_index=True)
```

# Get baseline for comparisons

```python
foo = df.groupby(["archetype", "baseline_transmission_metric", "transmission_level", "scenario_number"]).agg("mean").reset_index()
foo2 = df.groupby(["archetype", "baseline_transmission_metric", "transmission_level", "scenario_number"]).agg("std").reset_index()
foo3 = pd.merge(foo,foo2, suffixes=('','_STDEV'), on=["archetype", "baseline_transmission_metric", "transmission_level", "scenario_number"])

# foo_median = df.groupby(["archetype", "baseline_transmission_metric", "transmission_level", "scenario_number"]).agg("median").reset_index()
# foo3 = pd.merge(foo3, foo_median, suffixes=('','_MEDIAN'), on=["archetype", "baseline_transmission_metric", "transmission_level", "scenario_number"])
```

```python
base = foo3[foo3["scenario_number"]==30]
base = base[["archetype", 
             "baseline_transmission_metric",
             "transmission_level",
             "cases_treated", 
             'clinical_incidence0_5', 
             'clinical_incidence2_10', 
             'clinical_incidence6_15', 
             'clinical_incidence16_500',
             'clinical_incidence_all',
            'pfpr0_5',
            'pfpr2_10',
            'pfpr6_15',
             'pfpr16_500',
            'pfpr_all']]
base
```

```python
bar = pd.merge(foo3, base, how="left", on=["archetype", "baseline_transmission_metric", "transmission_level"], suffixes=("","_BASELINE"))
bar
```

```python
# Add total clinical burden fields:
case_type = "clinical"
for agebin in ["0_5", "2_10","6_15","16_500","_all"]:
    # for case_type in ["clinical", "severe"]:
    old = bar["{}_incidence{}_BASELINE".format(case_type, agebin)]
    new = bar["{}_incidence{}".format(case_type, agebin)]
    frac_averted = (old-new)/old

    bar["frac_averted_{}_{}".format(case_type, agebin)] = frac_averted
    
    # errorbar
    new_std = bar["{}_incidence{}_STDEV".format(case_type, agebin)]
    frac_averted_stdev = new_std/old
    
    bar["frac_averted_{}_{}_STDEV".format(case_type, agebin)] = frac_averted_stdev
```

```python
# Add prevalence reduction field:
for agebin in ["0_5", "2_10","6_15","16_500","_all"]:
    # for case_type in ["clinical", "severe"]:
    old = bar["pfpr{}_BASELINE".format(agebin)]
    new = bar["pfpr{}".format(agebin)]
    frac_averted = (old-new)/old

    bar["frac_averted_{}_{}_PFPR".format(case_type, agebin)] = frac_averted
    
    
    # errorbar
    new_std = bar["pfpr{}_STDEV".format(agebin)]
    frac_averted_stdev = new_std/old
    
    bar["frac_averted_{}_{}_PFPR_STDEV".format(case_type, agebin)] = frac_averted_stdev
```

# Add scenario information

```python
df_scenarios = pd.read_csv("../../run_sims/Assets/scenario_master_list.csv")
df_scenarios
```

```python
df_full = pd.merge(bar, df_scenarios, how="left", on=["archetype", 
                                                      "scenario_number"])
```
