```python
import pandas as pd
import numpy as np
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("talk")
sns.set_style("white")
```

```python
# Comparing calendars:

# DRC:
# Academic year period: From 5 September 2022 to 1st July 2023 included (222 business days)
# 1.	First trimester: Monday 5th September 2022 to Friday 23rd December 2022 included (92 business days). Holidays from 24th December 2022 to 7th January 2023
# 2.	Second trimester: Monday 9th January 2023 to Friday 1st April 2023 included (68 business days). Holidays from 3rd – 15th April 2023
# 3.	Third trimester: Monday 17th April 2023 to Friday 1st July 2023 included (62 business days).

t1 = [248, 	357]
t2 = [9, 91]
t3 = [107, 182]

drc = [t1,t2,t3]

for t in drc:
    plt.axvspan(t[0], t[1])
    
plt.xlim([0,365])
```

```python
# Cameroon:
# 5 Sept – 25 November – First Term 
# 28 Nov – 16 December – Second term first part 
# 3 Jan – 3 March – Second term  - second part 
# 6 Mar – 7 April – Third term – first part 
# 17 April – 2 Jun – Third term – second part 

t1 = [248, 329]
t2a = [332, 350]
t2b = [3, 62]
t3a = [65, 97]
t3b = [107, 153]

cameroon = [t1,t2a,t2b,t3a,t3b]

for t in cameroon:
    plt.axvspan(t[0], t[1])
```

```python
# Republic of Congo:
# Jan 4-Feb 24
# March 13-April 30
# May 15-July 3
# Sept 1-Oct 21
# Nov 3-Dec 18

t1 = [4,55]
t2 = [72,120]
t3 = [135,184]
t4 = [244,294]
t5 = [307,352]

roc = [t1,t2,t3,t4,t5]

for t in roc:
    plt.axvspan(t[0], t[1])
```

```python
for i,t in enumerate(drc):
    if i == 0:
        label = "DRC"
    else:
        label = None
    plt.axvspan(t[0], t[1], ymin=0,ymax=0.2, color="C0", label=label)
    
for i,t in enumerate(cameroon):
    if i == 0:
        label = "Cameroon"
    else:
        label = None
    plt.axvspan(t[0], t[1], ymin=0.2,ymax=0.4, color="C1", label=label)
    
for i,t in enumerate(roc):
    if i == 0:
        label = "RoC"
    else:
        label = None
    plt.axvspan(t[0], t[1], ymin=0.4, ymax=0.6, color="C2", label=label)
    

# Actual campaign days for IPTsc
for i, d in enumerate([60, 264, 349]):
    if i == 0:
        label = "Term campaigns"
    else:
        label = None
    plt.axvline(d, linestyle='dashed', color='black', label=label)
    
for i, d in enumerate([5, 35, 60, 90, 121, 151, 181, 244, 274, 305, 335]):
    if i == 0:
        label = "Month campaigns"
    else:
        label = None
    plt.axvline(d, linestyle='dotted', color='gray', label=label)

    
plt.yticks([])
plt.xlim([0,365])
# plt.ylim([0,1.8])
plt.legend(fontsize=12)
plt.title("School calendars")
plt.xlabel("Day")
```
