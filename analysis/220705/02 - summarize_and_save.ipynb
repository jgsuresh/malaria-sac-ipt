```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_context("talk")
sns.set_style("white")
```

```python
df = pd.read_csv("sim_data_full_220705.csv")
```

```python
foo = df.groupby(["archetype", "baseline_eir", "scenario_number"]).agg("mean").reset_index()
foo
```

```python
base = foo[foo["scenario_number"]==30]
base = base[["archetype", "baseline_eir", "cases_treated", 'clinical_incidence0_5', 'clinical_incidence2_10', 'clinical_incidence6_15', 'clinical_incidence16_500','clinical_incidence_all']]
base
```

```python
bar = pd.merge(foo, base, how="left", on=["archetype", "baseline_eir"], suffixes=("","_BASELINE"))
bar
```

```python
# Add total clinical burden fields:
case_type = "clinical"
for agebin in ["0_5", "2_10","6_15","16_500","_all"]:
    # for case_type in ["clinical", "severe"]:
    old = bar["{}_incidence{}_BASELINE".format(case_type, agebin)]
    new = bar["{}_incidence{}".format(case_type, agebin)]
    frac_averted = (old-new)/old

    bar["frac_averted_{}_{}".format(case_type, agebin)] = frac_averted
```

```python
bar
```

```python
df_cost = pd.read_csv("marita_results_2022-07-05.csv")
df_cost.drop(["Unnamed: 0", "drug_type", "Scenario", "cases"], axis=1, inplace=True)
df_cost
```

```python
df_save = pd.merge(bar, df_cost, how="left", on=["archetype", "baseline_eir", "scenario_number"])
df_save
```

```python
df_scenarios = pd.read_csv("../../run_sims/Assets/scenario_master_list.csv")
df = pd.merge(df_save, df_scenarios, on=["archetype", "scenario_number"])
df
```

```python
df.to_csv("sim_data_summary_220705.csv", index=False)
```
