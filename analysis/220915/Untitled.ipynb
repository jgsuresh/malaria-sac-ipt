```python
import pandas as pd
import numpy as np
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("talk")
sns.set_style("white")
```

```python
# df = pd.read_csv("sim_data_eb1c2efd-4b35-ed11-a9fc-b88303911bc1.csv")
# df = pd.read_csv("sim_data_07fb887e-5535-ed11-a9fc-b88303911bc1.csv")
# df

df1 = pd.read_csv("sim_data_07fb887e-5535-ed11-a9fc-b88303911bc1.csv")
df2 = pd.read_csv("sim_data_57f0bff1-e139-ed11-a9fc-b88303911bc1.csv")
d3 = pd.read_csv("sim_data_b1a1c23a-e639-ed11-a9fc-b88303911bc1.csv")
df = pd.concat([df1, df2, d3], ignore_index=True)
```

```python
df.columns
```

```python
sns.swarmplot(data=df,x="baseline_eir", y="clinical_incidence0_5", hue="smc_drug_config")
```

```python
for e, sdf in df.groupby("baseline_eir"):
    plt.figure()
    sns.swarmplot(data=sdf,x="smc_drug_config", y="clinical_incidence0_5")
    plt.title(f"Baseline EIR = {e}")
    plt.savefig(f"eir_{e}.png")
```
