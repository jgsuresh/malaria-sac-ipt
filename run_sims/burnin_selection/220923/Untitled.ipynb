```python
import pandas as pd
import numpy as np
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("talk")
sns.set_style("white")
```

```python
df = pd.read_csv("burnins.csv")
df
```

```python
df2 = pd.read_csv("burnin_endpoint_0dd75eef-9e3a-ed11-a9fc-b88303911bc1.csv")
foo = df2.groupby(["archetype", "habitat_scale"]).agg({"wet_season_rdt_prev": "mean"}).reset_index()
df_full = pd.merge(df,foo, how="left", on=["archetype", "habitat_scale"])
```

```python
df_pfpr = df_full[np.logical_not(np.isnan(df_full["burnin_approx_pfpr_2-10"]))].reset_index(drop=True)
df_pfpr
```

```python
df_eir = df_full[np.logical_not(np.isnan(df_full["burnin_approx_aeir"]))].reset_index(drop=True)
```

```python
sns.scatterplot(data=df_pfpr, x="burnin_approx_pfpr_2-10", y="annual_rdt_prev_SMOOTHED", hue="archetype")
```

```python
sns.scatterplot(data=df_pfpr, x="burnin_approx_pfpr_2-10", y="wet_season_rdt_prev", hue="archetype")
```

```python
sns.scatterplot(data=df_pfpr, x="burnin_approx_pfpr_2-10", y="aeir_SMOOTHED", hue="archetype")
plt.yscale("log")
```

```python
sns.scatterplot(data=df_eir, x="burnin_approx_aeir", y="pfpr_2-10_SMOOTHED", hue="archetype")
plt.xscale("log")
```
