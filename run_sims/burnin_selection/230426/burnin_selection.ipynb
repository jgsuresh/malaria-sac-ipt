```python
pfpr_selected = [0.01,0.05,0.1,0.2,0.3,0.4]
aeir_selected = [1,3,10,30,100]
```

```python
def select_closest(df):
    save_rows = []
    for p in pfpr_selected:
        i = np.argmin(np.abs(df["pfpr_2-10_SMOOTHED"]-p))
        s = df.iloc[i]
        s["burnin_approx_pfpr_2-10"] = p
        save_rows.append(s)

    for e in aeir_selected:
        i = np.argmin(np.abs(df["aeir_SMOOTHED"]-e))
        s = df.iloc[i]
        s["burnin_approx_aeir"] = e
        save_rows.append(s)

    return pd.DataFrame(save_rows)
```

```python
df_list = []

dc = select_closest(df_sahel)
dc["archetype"] = "Sahel"
df_list.append(dc.copy())

dc = select_closest(df_central)
dc["archetype"] = "Central"
df_list.append(dc.copy())

dc = select_closest(df_southern)
dc["archetype"] = "Southern"
df_list.append(dc.copy())


df_save = pd.concat(df_list).reset_index(drop=True)
```
